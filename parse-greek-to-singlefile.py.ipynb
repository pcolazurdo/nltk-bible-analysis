```python
import re
from os import listdir
from os.path import isfile, join

# r = '[\d+:\d+].+'
r = r'(?P<verse>\d+:\d+) (?P<text>\w+)'

mypath = 'data/greektext-textus-receptus/parsed/'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]

output = []
for f in onlyfiles:
    verse = ''
    with open(join(mypath, f)) as text:
        book_name = f.split('.')[0]
        for line in text.readlines():
#             print(line)
            regexp = re.match(r, line)
            if (regexp):
                output.append(verse + '\n')
                verse = book_name + ' ' + regexp.group('verse') + ' | ' + regexp.group('text')
            else:
                verse = verse + line
            verse = verse.rstrip("\n")
        output.append(verse)
    output.append(verse)
```

```python
with open ('output.txt', '+w') as out_file:
    for l in output:
        out_file.write(l)
    out_file.close()
```
