```python
from os import listdir
from os.path import isfile, join
import json
import itertools
```

```python
BIBLE_LOCATION='./Bible-kjv'
```

```python
files = ['{}/{}'.format(BIBLE_LOCATION,f) for f in listdir(BIBLE_LOCATION) if isfile(join(BIBLE_LOCATION, f)) and f.find('.json') >= 0]
```

```python
with open('{}/{}'.format(BIBLE_LOCATION,'Books.json')) as b:
    books = json.load(b)
```

```python
bible = {}
for file in books:
    file_name = '{}/{}.json'.format(BIBLE_LOCATION, file.replace(" ", ""))
    with open(file_name) as f:
        book = json.load(f)
        try:
            book_name = book['book']
            bible[book_name] = {'chapters': book['chapters']}
        except Exception:
            pass
```

```python
# Text is array of terms to search 
# but is array of terms to exclude
# books is array of books to search in
# Todo: add case sensitivity

def find_text(text, but=None, books=None):
    if books is not None:
        book_list = [ b for b in bible if b in books]
    else: 
        book_list = bible
    text = [text] if type(text) != list else text
#     but = [but] if type(but) != list  else but
    for book_name in book_list:
        for chapter_content in bible[book_name]['chapters']:
            for verse_content in chapter_content['verses']:
                for find_text in text:
                    if (verse_content['text'].find(find_text) >= 0):
                        found = True
                    else:
                        found = False
                        break
                not_found = True
                if but:
                    for not_in_text in but:
                        if (verse_content['text'].find(not_in_text) < 0):
                            not_found = True
                        else:
                            not_found = False
                            break
                if found and not_found:
                    yield(book_name, chapter_content['chapter'], verse_content['verse'], verse_content['text'])
```

```python
def count_coincidences(text, but=None):
    new_list = []
    l = list(find_text(text, but=but))
    l = [ a[0]  for a in l ]
    for k, g in itertools.groupby(sorted(l)):
        new_list.append( (k, len(list(g))) )
    return new_list
```

```python
# del (bible['Exodus'])
for x in find_text([ " four ", 'corners'], but=['thousand', 'hundred', 'year', 'twenty']):
    print ("{} {}:{} {}".format(x[0], x[1], x[2], x[3]))
sorted(count_coincidences([" four "], but=['thousand', 'hundred', 'year', 'twenty']), key=lambda x: x[1], reverse=True)
```
